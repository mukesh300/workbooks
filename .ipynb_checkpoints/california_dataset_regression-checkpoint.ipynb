```python
import warnings
warnings.simplefilter("ignore")

import pandas as pd
import numpy as np

import tensorflow as tf
from tensorflow import keras
from pandas_summary import DataFrameSummary
import seaborn as sns 

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
housing = fetch_california_housing()
housing.keys()
```

```python
X, y = housing.data, housing.target
print(f"input shape: {X.shape} {y.shape}\nInput features: {housing.feature_names}")
```

```python
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df["SalePrice"] = housing.target
dfs = DataFrameSummary(housing_df)
dfs.summary()
```

```python
housing_df["income_cat"] = pd.cut(housing_df["MedInc"], bins=[0., 1.5, 3.0, 4.5, 6., np.inf], labels=[1, 2, 3, 4, 5])
housing_df["income_cat"].hist()
```

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing_df, housing_df["income_cat"]):
    strat_train_set = housing_df.loc[train_index]
    strat_test_set = housing_df.loc[test_index]

for set_ in (strat_train_set, strat_test_set):
    set_.drop("income_cat", axis=1, inplace=True)
    
housing_df = strat_train_set.copy()
```

```python
grp_house_age = housing_df.groupby('HouseAge')
plt.figure(figsize=(15,9))
df_house_age = grp_house_age['SalePrice'].mean().reset_index()

sns.barplot(x=df_house_age.HouseAge, y=df_house_age.SalePrice, palette="Blues_d")
plt.xticks(rotation=90)
plt.title("SalePrice distribution based on age")
```

```python
from scipy.stats import norm, skew 

sns.set_style("whitegrid")
plt.figure(figsize=(12,4))

sns.distplot(housing_df.SalePrice, fit=norm);
(mu, sigma) = norm.fit(housing_df.SalePrice)
print( '\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f} )'.format(mu, sigma)],
            loc='upper right')

plt.ylabel('Frequency')
plt.title('SalePrice distribution')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.model_selection import cross_val_score

def display_scores(scores, algo):
    print(f"{algo} Scores: {scores}")
    print(f"{algo} Mean: {scores.mean()}")
    print(f"{algo} Standard deviation: {scores.std()}\n")

lin_reg = LinearRegression()
ridge_reg = Ridge(alpha=0.1, solver="sag", random_state=42)
lasso_reg = Lasso(alpha=0.1)
elastic_net = ElasticNet(alpha=0.1, l1_ratio=0.5, random_state=42)

lin_reg_rmse = np.sqrt(-cross_val_score(lin_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10))
ridge_reg_rmse = np.sqrt(-cross_val_score(ridge_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10))
lasso_reg_rmse = np.sqrt(-cross_val_score(lasso_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10))
elastic_net_rmse = np.sqrt(-cross_val_score(elastic_net, X_train, y_train, scoring="neg_mean_squared_error", cv=10)
                          )
display_scores(lin_reg_rmse, "lin_reg")
display_scores(ridge_reg_rmse, "ridge_reg")
display_scores(lasso_reg_rmse, "lasso_reg")
display_scores(elastic_net_rmse, "elastic_net")
```

```python
from sklearn.svm import LinearSVR, SVR

svm_reg = LinearSVR(epsilon=1.5, random_state=42)
svm_rbf_reg = SVR(kernel="rbf", C=100, gamma=5)

svm_reg.fit(X, y)
svm_rbf_reg.fit(X, y)

svm_reg_rmse = np.sqrt(-cross_val_score(svm_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10))
svm_rbf_reg_rmse = np.sqrt(-cross_val_score(svm_rbf_reg, X_train, y_train, scoring="neg_mean_squared_error", cv=10))

display_scores(svm_reg_rmse, "svm_reg")
display_scores(svm_rbf_reg_rmse, "svm_rbf_reg")
```
